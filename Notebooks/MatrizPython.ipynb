```python
#!/usr/bin/python
import time
import matplotlib.pyplot as plt
import numpy as np
T = 2

x = []
y = []

while T <= 5000:
    a = np.ones((T, T))
    b = np.ones((T, T))
    b+=1

    start_time = time.time()
    
    
    x.append(T)
    np.dot(a,b)
    y.append(time.time() - start_time)
    #print("--- %s seconds ---" % (time.time() - start_time))
    T+=100


plt.plot(x, y)
plt.show()
           
```
