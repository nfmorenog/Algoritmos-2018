```python
import time
import matplotlib.pyplot as plt


n = 0
x = []
y = []
def fibo(x):
    i = 1
    j = 0
    for x in range(0, x):
        j = i + j
        i = j - i        
    return j


while n < 200000:
    start_time = time.time()
    x.append(n)
    fibo(n)
    y.append(time.time() - start_time)
    n+=1000
   
   
    
	
    

plt.plot(x, y)
plt.xlabel('Fibonacci n')
plt.ylabel('Time')
plt.show()
```
